```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

import math
```

```python
scripts = []

for mask_idx in range(1):

    raster_plan = []

    X = Y = 32
    patch_dim = 1
    group_size = 4
    N = (X*Y) // group_size

    print(N)

    uni = tf.random.uniform(shape=(1, N))
    rand_indices = tf.argsort(uni, axis=-1)

    print(rand_indices.shape)

    mask_percentage = 0.1
    mask_num = int(N*mask_percentage)
    print(mask_num)

    mask_indices = rand_indices[:, :mask_num][0]
    unmask_indices = rand_indices[:, mask_num:][0]

    mask = np.zeros(shape=(X, Y), dtype=np.uint8)

    unravel_shape = (int(math.sqrt(N)), int(math.sqrt(N)))

    print(unravel_shape)

    counter = 0
    for item in np.transpose(np.unravel_index(mask_indices, unravel_shape, order='C')):
        x, y = item

        i = (x*2)
        j = (y*2)
        mask[i, j] = 1
        a = (i, j)
        b = (i+1, j)
        c = (i, j+1)
        d = (i+1, j+1)
        raster_plan.append([a, b, c, d, patch_dim, X, Y])


        i = (x*2)+1
        j = (y*2)
        mask[i, j] = 1
        a = (i, j)
        b = (i+1, j)
        c = (i, j+1)
        d = (i+1, j+1)
        raster_plan.append([a, b, c, d, patch_dim, X, Y])    


        i = (x*2)
        j = (y*2)+1
        mask[i, j] = 1
        a = (i, j)
        b = (i+1, j)
        c = (i, j+1)
        d = (i+1, j+1)
        raster_plan.append([a, b, c, d, patch_dim, X, Y])    


        i = (x*2)+1
        j = (y*2)+1
        mask[i, j] = 1
        a = (i, j)
        b = (i+1, j)
        c = (i, j+1)
        d = (i+1, j+1)
        raster_plan.append([a, b, c, d, patch_dim, X, Y])  
        


        counter += 1

    plt.imshow(mask)
    plt.xlabel('column axis')
    plt.ylabel('row axis')
    plt.show()

    import pandas as pd

    column_values = ['a', 'b', 'c', 'd', 'patch size', 'x', 'y']
    df = pd.DataFrame(
        data=raster_plan,
        columns = column_values
    )

    print('Coordinates for Masking of', (1.0-mask_percentage)*100, '%')

    df_orig = df
    df
    
#     x_left = -0.5
#     x_right = 0.5
    
#     y_up = 0.5
#     y_down = -0.5

#     x_left = -0.45
#     x_right = 0.45

#     y_up = 0.4
#     y_down = -0.5

    x_left = -0.35
    x_right = 0.35

    y_up = 0.2
    y_down = -0.5

    x_range = abs(x_left) + abs(x_right)
    y_range = abs(y_up) + abs(y_down)

    pixels = X

    x_step_size = x_range / pixels
    y_step_size = y_range / pixels


    x = np.arange(
        start=x_left, 
        stop=x_right+x_step_size, 
        step=x_step_size
    )

    y = np.arange(
        start=y_down, 
        stop=y_up+y_step_size, 
        step=y_step_size
    )

    conversion_x = {}
    for idx, elem in enumerate(x):
        conversion_x[idx] = elem

    conversion_y = {}
    for idx, elem in enumerate(y):
        conversion_y[idx] = elem

    conversion_x = {}
    for idx, elem in enumerate(x):
        conversion_x[idx] = round(elem, 10)

    conversion_y = {}
    for idx, elem in enumerate(y):
        conversion_y[idx] = round(elem, 10)


    raster_angle_plan = []

    for a, b, c, d, patch, x, y in raster_plan:
        a = (conversion_x[a[0]], conversion_x[a[1]])
        b = (conversion_x[b[0]], conversion_x[b[1]])
        c = (conversion_x[c[0]], conversion_x[c[1]])
        d = (conversion_x[d[0]], conversion_x[d[1]])

        raster_angle_plan.append([a, b, c, d, patch, x, y])


    import pandas as pd

    column_values = ['a', 'b', 'c', 'd', 'patch size', 'x', 'y']
    df = pd.DataFrame(
        data=raster_angle_plan,
        columns = column_values
    )

    # print('Angle Coordinates for Masking of', mask_percentage*100, '%')
    # df

    raster_angle_small_plan = []

    for a, b, c, d, patch, x, y in raster_plan:

        x_small = conversion_x[a[0]]
        x_big = conversion_x[b[0]]
        y_small = conversion_y[a[1]]
        y_big = conversion_y[c[1]]
        s = x_step_size
        u = y_step_size

        raster_angle_small_plan.append([
            x_small,
            x_big,
            y_small,
            y_big,
            s,
            u,
            int(a[0]),
            int(a[1]),
            patch_dim
        ])

    column_values = ['x', 'X', 'y', 'Y', 's', 'u', 'origin_x', 'origin_y', 'patch_dim']
    df = pd.DataFrame(
        data=raster_angle_small_plan,
        columns = column_values
    )

    df = df.round(10)

    for index, row in df.iterrows():
        x, X, y, Y, s, u, origin_x, origin_y, patch_dim = row
        origin_x = int(origin_x)
        origin_y = int(origin_y)
        patch_dim = int(patch_dim)
        script = f'python2 ../Direct.py -x {x} -X {X} -y {y} -Y {Y} -s {s} -u {u} -v -t 10000 -f ../MAE-Windows/{origin_x}-{origin_y}-{patch_dim}-{int(100*(1.0-mask_percentage))}-{mask_idx}.csv'
        scripts.append(script)


file_path = "Direct-scripts.txt"

with open(file_path, "w") as file:
    for item in scripts:
        file.write(item + "\n")

file.close()
```
