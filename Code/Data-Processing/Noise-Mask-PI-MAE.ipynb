```python
import os
import cv2
import random
import string
import numpy as np
import matplotlib.pyplot as plt
```

```python
masked_images_path = 'Jan-MAE-Windows/SHOT-NOISE/GOOD-DATA/90mask-good/'
dest_path = 'Jan-MAE-Windows/SHOT-NOISE/NOISE-MASK/90mask-noise/'

counter = 0
for fn in os.listdir(masked_images_path):
    if 'cv2' in fn:
        counter += 1
        print(fn)
        
        img_path = os.path.join(masked_images_path, fn)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        
        obj, mask_percent, _, idx, _ = fn.split('.')[0].split('-')
        
        mask_fn = f'{obj}-{mask_percent}-1-{idx}-plt-mask.png'
        mask_path = os.path.join(masked_images_path, mask_fn)
        print(mask_path)
        
        noisy = np.copy(img)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        not_mask = ~mask
        x, y = np.where(not_mask == 0)
        for i in range(x.shape[0]):
            noisy[x[i], y[i]] = random.uniform(0, 254)
        
        fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))
        
        ax[0].imshow(mask, cmap='gray')
        ax[0].set_title("Mask")

        ax[1].imshow(img, cmap='gray')
        ax[1].set_title("Image")
        
        ax[2].imshow(noisy, cmap='gray')
        ax[2].set_title("Noisy")
        
        plt.show()
        
# Y-75-1-23-cv2.png
# Y-75-1-16-plt-mask.png
        
        noisy_fn = f'{obj}-{mask_percent}-1-{idx}-cv2.png'
        noisy_dest = os.path.join(dest_path, noisy_fn)
        cv2.imwrite(noisy_dest, noisy)
        
        mask_fn = f'{obj}-{mask_percent}-1-{idx}-plt-mask.png'
        mask_dest = os.path.join(dest_path, mask_fn)
        cv2.imwrite(mask_dest, mask)
        
        
counter == (25 + 10)
```
